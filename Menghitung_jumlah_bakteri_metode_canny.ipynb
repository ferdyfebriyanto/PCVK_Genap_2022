```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/drive/MyDrive/PCVK_MM/TugasProject/Bacteria/001.png')

#gambar di warna asli
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
#menunjukkan gambar pada warna gray
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2GRAY), cmap='gray')
```

```python
#membuat sebuah gambar blur
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

blur = cv.GaussianBlur(gray, (11,11), 1)
plt.imshow(blur, cmap='gray')
```

```python
canny = cv.Canny(blur, 60, 150, 3)
plt.imshow(canny, cmap= 'gray')
```

```python
dilated = cv.dilate(canny, (1,1), iterations = 1)
plt.imshow(dilated, cmap='gray')
```

```python
(cnt, hierarchy) = cv.findContours(dilated.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.drawContours(rgb, cnt, -1, (0,255,0),3)
plt.figure(figsize=(20,10))
plt.imshow(rgb)
```

```python
print("Bakteri berjumlah: ")
len(cnt)
```
